```python
import os
from datetime import datetime
```

```python
# Classe para formatar a função print
class color:
    PURPLE = '\033[95m'
    CYAN = '\033[96m'
    DARKCYAN = '\033[36m'
    BLUE = '\033[94m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    ORANGE = '\033[33m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    END = '\033[0m'
```

```python
# Método que retorna o diretório de Datasets
def setDatasetPath():
    return 'D:\DataScience\Bootcamp-Analista-de-Dados-IGTI\Modulo-03\Exercícios'

# Método para leitura de arquivo (Lê todo o conteúdo de uma vez)
def readFileContent(filename):
    try:
        fileobject = open(filename, 'r', encoding='utf-8')
        print(f'{color.BOLD + color.BLUE}\nO conteúdo do arquivo {filename} é: \n{color.END}')
        if fileobject.mode == 'r':
            filecontents = fileobject.read()
            print(filecontents)
        fileobject.close()
    except FileNotFoundError:
        print(f'{color.BOLD + color.RED}\nERRO AO TENTAR LER O ARQUIVO! \nO Arquivo' 
              f'{filename} não existe!{color.END}')
            
# Método para leitura de arquivo, lê as linhas e as insere em uma lista
def readFileLines(filename):
    try:
        print(f'{color.BOLD + color.BLUE}\nAs linhas do arquivo {filename} são:\n{color.END}')
        fileobject = open(filename, 'r', encoding='utf-8')
        filelines = fileobject.readlines()
        
        # Variável de apoio para controle do número de linha
        numline = 1
        
        # Para cada linha da lista imprime o valor da linha
        # end='' exclui a quebra de linha
        for line in filelines:
            print(f'Linha {str(numline)}: {line}', end='')
            numline += 1
            
        fileobject.close()
    except FileNotFoundError:
        print(f'{color.BOLD + color.RED}\nERRO AO TENTAR LER O ARQUIVO! \nO Arquivo' 
              f'{filename} não existe!{color.END}')

# Método para alteração do arquivo
def appendFile(filename):
    # Abri arquivo em modo de alteração (append)
    # "a" - append = inclui conteúdo ao final do arquivo
    # cria um arquivo se o próximo especificado não existir
    fileobject = open(filename, 'a', encoding='utf-8')
    
    # Adiciona quebra de linha
    fileobject.write('\n') 
    
    # Adiciona uma linha informando que é o final do arquivo
    fileobject.write('Olá pessoal!')
    now = datetime.now() 
    date_time = now.strftime('%d/%m/%Y, %H:%M:%S')
    fileobject.write(f'\n--> Hoje é {str(date_time)}')
    
    fileobject.write('\n\n')
    fileobject.write('Olá pessoal!')
    fileobject.write(f'\n--> Hoje é {str(date_time)}')
    fileobject.close()
    
# Método para escrita do arquivo
def writeFile(filename):
    # Abri o arquivo em modo de escrita (write)
    # "w" - Write - substui qualquer conteúdo existente
    # cria um arquivo se o arquivo especificado não existir
    
    fileobject = open(filename, 'w', encoding='utf-8')
    fileobject.write(f'Abri o arquivo {filename} em modo de escrita (w)!')
    
    now = datetime.now()
    date_time = now.strftime('%d/%m/%Y, %H:%M:%S')
    fileobject.write(f'\n--> Hoje é {str(date_time)}')
    fileobject.close()

# Método para criação de arquivo
def createFile(filename):
    # Cria o arquivo usando o método open()
    # "x" - cria o arquivo e retorna erro caso o arquivo já exista
    # filename = setDatasetPath() + \escritorCreate1.txt
    
    try:
        fileobject = open(filename, 'x', encoding='utf-8')
        fileobject.write('Este é um novo arquivo criado usando o modo de criação (x).')
        fileobject.close()
    except FileExistsError:
        print(f'\nERRO AO TERNTAR CRIAR O ARQUIVO! \nO arquivo {filename} já existe!')
    
```

```python
# Movendo arquivo de um diretório para o outro
import shutil

folderSource = 'D:\\DataScience\\Bootcamp-Analista-de-Dados-IGTI\\Modulo-03\\Exercícios\\source'
folderTarget = 'D:\\DataScience\\Bootcamp-Analista-de-Dados-IGTI\\Modulo-03\\Exercícios\\target'

sourceList = os.listdir(folderSource)
targetList = os.listdir(folderTarget)

print(f'{color.BOLD+color.PURPLE}Antes de tentar mover os arquivos!\n{color.END}')

print('Arquivos na origem', sep='')
print(sourceList)
print('Arquivos no destino', sep='')
print(targetList)

# Move conteúdo da pasta source para a pasta target
# shutil.move(folderSource, folderTarget)

for filename in sourceList:
    try:
        if os.path.isfile(os.path.join(folderSource, filename)):
            filepath = os.path.join(folderSource, filename)
            shutil.move(filepath, folderTarget)
    except Exception as e:
            print(f'Falha ao mover {filepath}. Razão da falha: {e}')
            
sourceList = os.listdir(folderSource)
targetList = os.listdir(folderTarget)
print('\n')
print(f'{color.BOLD+color.GREEN}Após tentar mover os arquivos!\n{color.END}')
print('Arquivos na origem', sep='')
print(sourceList)
print('Arquivos no destino', sep='')
print(targetList)
```

```python
# Copiando arquivo de um diretório para o outro
import shutil

folderSource = 'D:\\DataScience\\Bootcamp-Analista-de-Dados-IGTI\\Modulo-03\\Exercícios\\source2'
folderTarget = 'D:\\DataScience\\Bootcamp-Analista-de-Dados-IGTI\\Modulo-03\\Exercícios\\target'

sourceList = os.listdir(folderSource)
targetList = os.listdir(folderTarget)

print(f'{color.BOLD+color.PURPLE}Antes de tentar mover os arquivos!\n{color.END}')

print('Arquivos na origem', sep='')
print(sourceList)
print('Arquivos no destino', sep='')
print(targetList)

# Move conteúdo da pasta source para a pasta target
# shutil.move(folderSource, folderTarget)

for filename in sourceList:
    try:
        if os.path.isfile(os.path.join(folderSource, filename)):
            filepath = os.path.join(folderSource, filename)
            filepath2 = os.path.join(folderTarget, filename)
            shutil.copy(filepath, filepath2)
    except Exception as e:
            print(f'Falha ao copiar {filepath}. Razão da falha: {e}')
            
sourceList = os.listdir(folderSource)
targetList = os.listdir(folderTarget)
print('\n')
print(f'{color.BOLD+color.GREEN}Após tentar mover os arquivos!\n{color.END}')
print('Arquivos na origem', sep='')
print(sourceList)
print('Arquivos no destino', sep='')
print(targetList)
```
